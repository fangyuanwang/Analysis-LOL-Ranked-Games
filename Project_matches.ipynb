```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

```python
matches_df = pd.read_csv("projectData/matches.csv")
matches_df.head(10)
```

```python
matches_average_duration_by_version = matches_df.groupby('version').mean()
matches_average_duration_by_version = matches_average_duration_by_version.drop(['id','gameid','queueid','seasonid','creation'],axis=1)
matches_average_duration_by_version.head(10)
```

```python
matches_average_duration_by_version.plot(figsize=(20,10))
plt.title('Average Duration of Match Based On Each Version in Seconds')
```

```python
matches_average_duration_by_season = matches_df.groupby('seasonid').mean()
matches_average_duration_by_season = matches_average_duration_by_season.drop(['id','gameid','queueid','creation'],axis=1)
matches_average_duration_by_season.head(10)
```

```python
matches_average_duration_by_season.plot(figsize=(10,5))
plt.title('Average Duration of Match Based On Each Season in Seconds')
```
